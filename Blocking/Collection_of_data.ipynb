# Data Writer

This script is used to save ERA-Interim 500 hPa data for the djf of certain years, it has four data points per day and it may be altered to give less data. It was designed to run from a JASMIN jupyter notebook. However, a daymean 2.5x2.5 500 hpa geopotential height for the region will run in the TM90 codebook.

```python

#need to use {var} in some 
#constructing yearly path:
for year in np.arange(1980,2018,1):
    pathlist= list()
    string1 = str(year) + '/01'
    string2 = str(year) + '/02'
    string3 = str(year-1) + '/12'
    pathlist.append(string1)
    pathlist.append(string2)
    pathlist.append(string3)
    !mkdir data/{year}
    for path in pathlist:
        netcdf_name = path.replace('/','_')
        !cdo select,name=Z /badc/ecmwf-era-interim/data/gg/ap/{path}/*/*.nc ~/Atmospheric_blocking/data/{year}/{netcdf_name}.nc

    !cdo select,name=Z ~/Atmospheric_blocking/data/{year}/*.nc ~/Atmospheric_blocking/data/{year}djf_trial.nc
    !cdo -remapbil,~/Atmospheric_blocking/data/grid_template.txt ~/Atmospheric_blocking/data/{year}djf_trial.nc ~/Atmospheric_blocking/data/{year}djf_final.nc
    !rm -r ~/Atmospheric_blocking/data/{year}
    !rm ~/Atmospheric_blocking/data/{year}djf_trial.nc
    

    
```
